```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df["gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.show()
```

```python
df["race/ethnicity"].value_counts().plot(kind="bar")

plt.title("Race/Ethnicity")
plt.show()
```

```python
df["parental level of education"].value_counts().plot(kind="bar")

plt.title("Parental Education")
plt.xticks(rotation=45)
plt.show()
```

```python
df["lunch"].value_counts().plot(kind="bar")

plt.title("Lunch Type")
plt.show()
```

```python
sns.histplot(df["math score"], kde=True)

plt.title("Math Score Distribution")
plt.show()
```

```python
sns.histplot(df["reading score"], kde=True)

plt.title("Reading Score Distribution")
plt.show()
```

```python
sns.histplot(df["writing score"], kde=True)

plt.title("Writing Score Distribution")
plt.show()
```

```python
sns.boxplot(x=df["math score"])

plt.title("Math Score Boxplot")
plt.show()
```

```python
sns.boxplot(x=df["reading score"])

plt.title("Reading Score Boxplot")
plt.show()
```

```python
sns.boxplot(x=df["writing score"])

plt.title("Writing Score Boxplot")
plt.show()
```

```python
corr = df[["math score","reading score","writing score"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x="gender", y="math score", data=df)

plt.title("Gender vs Math Score")
plt.show()
```

```python
sns.boxplot(x="gender", y="reading score", data=df)

plt.title("Gender vs Reading Score")
plt.show()
```

```python
sns.boxplot(x="gender", y="writing score", data=df)

plt.title("Gender vs Writing Score")
plt.show()
```

```python
sns.boxplot(x="lunch", y="math score", data=df)

plt.title("Lunch vs Math Score")
plt.show()
```

```python
sns.boxplot(x="test preparation course", y="math score", data=df)

plt.title("Test Preparation vs Math Score")
plt.show()
```

```python
sns.boxplot(x="test preparation course", y="reading score", data=df)

plt.title("Test Preparation vs Reading Score")
plt.show()
```

```python
sns.boxplot(x="test preparation course", y="writing score", data=df)

plt.title("Test Preparation vs Writing Score")
plt.show()
```
